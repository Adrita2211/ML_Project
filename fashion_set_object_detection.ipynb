```python
import ultralytics
from ultralytics import YOLO
model = YOLO('yolov8m.pt')
#model =YOLO('yolov8n-seg.pt')
```

```python
fashion_model = model.train(data='/content/Clothing-Detection-1/data.yaml',seed=42,epochs=5,batch=16)
```

```python
results = model.predict("/content/Clothing-Detection-1/valid/images/test_img_fashion.png")
```

```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

for i, result in enumerate(results):
    original_image = result.orig_img  # BGR format

    for j, box in enumerate(result.boxes):
        # Get bounding box coordinates
        x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())

        # Crop the object
        cropped_object = original_image[y1:y2, x1:x2]

        # Get class name and confidence
        cls_id = int(box.cls[0])
        cls_name = result.names[cls_id]
        conf = float(box.conf[0])

        # Convert BGR to RGB for matplotlib
        cropped_rgb = cv2.cvtColor(cropped_object, cv2.COLOR_BGR2RGB)

        # Display the cropped object
        plt.figure(figsize=(5, 5))
        plt.imshow(cropped_rgb)
        plt.title(f'{cls_name} (Confidence: {conf:.2f})')
        plt.axis('off')
        plt.show()

        # Save the cropped object
        cv2.imwrite(f'img_{i}_obj_{j}_{cls_name}_{conf:.2f}.jpg', cropped_object)
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
for i, result in enumerate(results):
    if result.masks is not None:
        for j, (box, mask) in enumerate(zip(result.boxes, result.masks)):
            x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
            mask = mask.data[0].cpu().numpy()
            mask = cv2.resize(mask, (x2-x1, y2-y1))

            # Create 4-channel image (BGRA)
            cropped = cv2.cvtColor(result.orig_img[y1:y2, x1:x2], cv2.COLOR_BGR2BGRA)

            # Apply mask to alpha channel
            cropped[:, :, 3] = (mask * 255).astype(np.uint8)

            cv2.imwrite(f'transparent_{i}_{j}.png', cropped)
            # Display the cropped object
            plt.figure(figsize=(5, 5))
            plt.imshow(cropped)
            plt.title(f'{cls_name} (Confidence: {conf:.2f})')
            plt.axis('off')
            plt.show()
```
