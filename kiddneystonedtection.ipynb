```python
# Step 2: Load the first 5 images from the test folder
import os
test_folder = '/content/drive/MyDrive/Dataset/test/images'  # Replace with your test folder path
image_files = sorted(os.listdir(test_folder))[:5]  # Get the first 5 images
image_paths = [os.path.join(test_folder, img) for img in image_files]
```

```python
# Step 3: Run inference on the images
results = model.predict(image_paths)  # Run predictions on the images
```

```python
import cv2
import matplotlib.pyplot as plt
# Step 4: Display predictions using YOLO's built-in methods
# Step 4: Resize and display all 5 images in a single window
plt.figure(figsize=(15, 10))  # Set the figure size

for i, result in enumerate(results):
    # Get the plotted image with bounding boxes
    plotted_image = result.plot()  # Returns a numpy array (BGR format)

    # Convert BGR to RGB for matplotlib
    plotted_image_rgb = cv2.cvtColor(plotted_image, cv2.COLOR_BGR2RGB)

    # Add subplot for each image
    plt.subplot(2, 3, i + 1)  # 2 rows, 3 columns (adjust as needed)
    plt.imshow(plotted_image_rgb)
    plt.title(f'Image {i + 1}')
    plt.axis('off')  # Hide axes

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()  # Display the figure
```
